```python
from bertopic import BERTopic
from transformers import AutoModel, AutoTokenizer
import torch
import pandas as pd
from sentence_transformers import SentenceTransformer
import nltk
import numpy as np
import random
import spacy

nltk.download('punkt_tab')
```

```python
nlp = spacy.load("en_core_web_sm")  #consider:en_core_web_lg
```

```python
model = SentenceTransformer('AI-Growth-Lab/PatentSBERTa')
```

```python
#Embedding loop including sentence tokenization

def tokenize_sentences(text, nlp):
    doc = nlp(text)
    sentences = [sent.text for sent in doc.sents]
    return sentences

def generate_embeddings(sentences, model):
    embeddings = model.encode(sentences, convert_to_numpy=True)
    return embeddings

def chunk_sentence(sentence, max_length=512, overlap=50):
    tokens = sentence.split()
    chunks = []
    for i in range(0, len(tokens), max_length - overlap):
        chunk = " ".join(tokens[i:i + max_length])
        chunks.append(chunk)
    return chunks

all_embeddings = []
all_application_numbers = []
all_patentees = []
all_sentences = []

for index, row in data.iterrows():
    patent_claim = row["Claims"]
    application_number = row["ApplicationNumber"]
    patentee = row["ProbablePatentAssignee"]

    # Tokenization
    sentences = tokenize_sentences(patent_claim, nlp)

    # Chunking
    for sentence in sentences:
        if len(sentence.split()) > 512:
            chunks = chunk_sentence(sentence)
            for chunk in chunks:
                embeddings = generate_embeddings([chunk], model)
                all_embeddings.extend(embeddings)
                all_application_numbers.extend([application_number])
                all_patentees.extend([patentee])
                all_sentences.extend([chunk])
        else:
            embeddings = generate_embeddings([sentence], model)
            all_embeddings.extend(embeddings)
            all_application_numbers.extend([application_number])
            all_patentees.extend([patentee])
            all_sentences.extend([sentence])

# creating df to save embeddings
results_df = pd.DataFrame({
    "applicationnumber": all_application_numbers,
    "patentee": all_patentees,
    "sentence_chunk": all_sentences,
    "embedding": all_embeddings
})

np.save("patent_sentence_embeddings_chunked.npy", np.array(results_df['embedding'].tolist()))
results_df.to_csv("patent_sentence_embeddings_chunked_metadata.csv", index=False)

print("Chunked embeddings and metadata saved.")
```
