```python
import spacy
from scispacy.abbreviation import AbbreviationDetector
```

```python
nlp = spacy.load("en_core_sci_lg")
nlp.add_pipe("abbreviation_detector")
```

## Creating abbreviation(s) dictionary

```python
file_path = "/content/drive/MyDrive/Theses/final_patent_text.csv"
```

```python
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python
stopwords = ["i", "ii", "iii", "iv", "v", "vi", "vii", "viii", "ix", "x", "xi", "xii", "xiii", "xiv", "xv", "xvi", "xvii", "xviii", "xix", "xx", "xxi", "xxii", "xxiii", "xxiv", "xxv", "xxvi", "xxvii", "xxviii", "xxix", "xxx"]

unique_abbreviations = {}

for claim_text in data['Claims'].iloc[:].astype(str):
    doc = nlp(claim_text)
    for abrv in doc._.abbreviations:
        if abrv.text not in stopwords:
            unique_abbreviations[abrv.text] = abrv._.long_form.text

abbreviations_df = pd.DataFrame(list(unique_abbreviations.items()), columns=["Abbreviation", "Definition"])
abbreviations_df.to_csv('unique_abbreviations.csv', index=False)
```

```python
df = pd.read_csv('unique_abbreviations.csv')
```
